```python
import os
import json
import matplotlib.pyplot as plt
import pandas as pd
from tqdm import tqdm
```

```python
meta_file_path = os.path.join('/mnt/disks/sdb/dataset', "song_meta.json")
if os.path.isfile(meta_file_path):
    with open(meta_file_path) as f:
        song_meta = json.load(f)
```

```python
genre_key = {}
for song in song_meta:
    for genre in song['song_gn_gnr_basket']:
        try:
            genre_key[genre] += 1
        except:
            genre_key[genre] = 1
```

```python
genre_key
```

```python
adjust_meta = []
for song in tqdm(song_meta):

    if 'GN1200' in song['song_gn_gnr_basket']: #  'GN1200': '랩/힙합',  → 'GN0300'
        song['song_gn_gnr_basket'].remove('GN1200')
        song['song_gn_gnr_basket'].append('GN0300')
        
    if 'GN2700' in song['song_gn_gnr_basket']: #  'GN2700': 'EDM', → 'GN1100'
        song['song_gn_gnr_basket'].remove('GN2700') 
        song['song_gn_gnr_basket'].append('GN1100')
        
    if 'GN2600' in song['song_gn_gnr_basket']: #  'GN2600': '일렉트로니카', → 'GN1100'
        song['song_gn_gnr_basket'].remove('GN2600')
        song['song_gn_gnr_basket'].append('GN1100')
        
    if 'GN0600' in song['song_gn_gnr_basket']: #  'GN0600': '록/메탈', → 'GN1000'
        song['song_gn_gnr_basket'].remove('GN0600')
        song['song_gn_gnr_basket'].append('GN1000')
        
    if 'GN0400' in song['song_gn_gnr_basket']: #  'GN0400': 'R&B/Soul', → 'GN1300'
        song['song_gn_gnr_basket'].remove('GN0400')
        song['song_gn_gnr_basket'].append('GN1300')
        
    if 'GN0800' in song['song_gn_gnr_basket']: #  'GN0500':'포크/블루스’, → 'GN1400'
        song['song_gn_gnr_basket'].remove('GN0800')
        song['song_gn_gnr_basket'].append('GN1400')
    
    if 'GN1900' in song['song_gn_gnr_basket']: # 'GN1900': 'J-POP', 
        song['song_gn_gnr_basket'].remove('GN1900')
        
        if "GN1902" in song['song_gn_dtl_gnr_basket']: # "GN1902": "POP", → 'GN0900'
            song['song_gn_gnr_basket'].append('GN0900')
            
        if "GN1903" in song['song_gn_dtl_gnr_basket']: # "GN1903": "록", → 'GN1000'
            song['song_gn_gnr_basket'].append('GN1000')
            
        if "GN1904" in song['song_gn_dtl_gnr_basket']: # "GN1904": "일렉트로니카", → 'GN1100'
            song['song_gn_gnr_basket'].append('GN1100')
 
        if "GN1905" in song['song_gn_dtl_gnr_basket']: # "GN1905": "랩/힙합", → 'GN0300'
            song['song_gn_gnr_basket'].append('GN0300')
            
        if "GN1906" in song['song_gn_dtl_gnr_basket']: # "GN1906": "R&B/Soul", → 'GN1300'
            song['song_gn_gnr_basket'].append('GN1300')
      
        if "GN1908" in song['song_gn_dtl_gnr_basket']: # "GN1908": "뉴에이지", → 'GN1800'
            song['song_gn_gnr_basket'].append('GN1800') 
            
        if "GN1909" in song['song_gn_dtl_gnr_basket']: # "GN1909": "재즈", → 'GN1700'
            song['song_gn_gnr_basket'].append('GN1700')    

    adjust_meta.append(song)
```

```python
final_meta = []
for song in tqdm(adjust_meta):
    if 'GN2500' in song['song_gn_gnr_basket'] and len(song['song_gn_gnr_basket']) > 1:
        song['song_gn_gnr_basket'].remove('GN2500') #  'GN2500': '아이돌', 
        
    if 'GN0500' in song['song_gn_gnr_basket'] and len(song['song_gn_gnr_basket']) > 1:
        song['song_gn_gnr_basket'].remove('GN0500') #  'GN0500': '인디음악', 
        
    if 'GN1500' in song['song_gn_gnr_basket'] and len(song['song_gn_gnr_basket']) > 1:
        song['song_gn_gnr_basket'].remove('GN1500') #  'GN1500': 'OST',  
    
    final_meta.append(song)
```
